```python
#first import pandas as pd
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#creating data frame
df = pd.read_csv('smartphones.csv')
```

```python
#as per task we using info function in data frame
df.info('smartphones.csv')
```

```python
#as per task we are using the describe function in data frame
df.describe(include='all')
```

```python
#as per task we are using head function in data frame
df.head(3)
```

```python
#as per task we are going to find the missing values
missingvalues = df.isnull().sum()
print("\nMissing values")
print(missingvalues)
```

```python
#as per task we are deleting the duplicates
df = df.drop_duplicates()
print("Duplicates are removed")
```

```python
df.dtypes
```

```python
#as per task we are changing the price datatypes object to int
df['price'] = df['price'].astype(str)
df['price'] = df['price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df.dtypes
```

```python
#as per task we are changing the ram data type object to float
df['ram'] = df['ram'].str.extract('(\d+)').astype(float)
df.dtypes
```

```python
#as per task we are changing the display data type object to float
df['display'] = df['display'].str.extract('([\d.]+)').astype(float)
df.dtypes
```

```python
#as per task we are changing the battery data type object to float
df['battery'] = df['battery'].str.extract('(\d+)').astype(float)
df.dtypes
```

```python
#as per task we are changing the camera data type object to float
df['camera'] = df['camera'].str.extract(r'(\d+)\s*MP').astype(float)
df.dtypes
```

```python
df.info()
```

```python
df.head()
```

```python
# Set the style for the plots
sns.set_style("whitegrid")

# Create the histogram for the 'price' column
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')

# Add titles and labels
plt.title('Distribution of Mobile Phone Prices', fontsize=16)
plt.xlabel('Price (in Currency Units)', fontsize=12)
plt.ylabel('Frequency (Count of Phones)', fontsize=12)

# Display the plot
plt.show()
```

```python
# Create the scatter plot for 'price' vs. 'rating'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='rating', data=df, hue='rating', palette='viridis', size='rating', sizes=(50, 200), legend=False)

# Add titles and labels
plt.title('Relationship between Price and Rating', fontsize=16)
plt.xlabel('Price (in Currency Units)', fontsize=12)
plt.ylabel('Rating', fontsize=12)

# Add a text annotation for context
plt.text(df['price'].min(), df['rating'].min(), 'Each point is a phone model.', fontsize=10, alpha=0.7)

# Display the plot
plt.show()
```
